```python
import pandas as pd
df = pd.read_csv('titanic.csv')
# 2.Carga el dataset en un DataFrame de pandas
print("\n===================")
print(" CARGA DEL DATASET")

display(df.head(10))

# 1.Analisis Estadistica de los datos
print("\n=======================")
print(" Analisis estadisticos\n")
display(df.describe())

# 3. Calcular el procentaje de supervivencia de los pasajeros

print("\n=============================")
print(" Porcentaje de supervivencia\n")
supervivencia_tasa = df['Survived'].mean()*100
print(f"Sobrevivio el {supervivencia_tasa:.3f}%")

# 3. Calcular el porsentaje de pasajeros segun cada clase

print("\n========================================")
print(" Porcentaje de pasajeros segun la clase\n")
clase_tasa = df['Pclass'].value_counts(normalize=True)*100
print(clase_tasa.round(3))

# 4. Determinar cuantos hombres y mujeres viajaron al titanic

print("\n========================================")
print(" Mujeres y Hombres que viajaron en el titanic\n")
viaje_HyM = df['Sex'].value_counts()
print (viaje_HyM)
```
